```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation
from matplotlib import colors
from matplotlib import quiver
import math

# To show plots
%matplotlib inline

# Reload notbook when changes are made
%reload_ext autoreload
%autoreload 2
```

### Results at the end of the 1st iteration (iteration 0)

```python
velX = pd.read_csv('velocity_X_0.txt', sep = ',')
velX = velX.sort_index(ascending=False)
velX.set_index('y_coord', inplace=True)

velY = pd.read_csv('velocity_Y_0.txt', sep = ',')
velY = velY.sort_index(ascending=False)
velY.set_index('y_coord', inplace=True)

p = pd.read_csv('pressure_0.txt', sep = ',')
p = p.sort_index(ascending=False)
p.set_index('y_coord', inplace=True)

vel_mag = pd.read_csv('velocity_Magnitude_0.txt', sep = ',')
vel_mag = vel_mag.sort_index(ascending=False)
vel_mag.set_index('y_coord', inplace=True)

velX_normalized = pd.read_csv('velocity_X_Normalized_0.txt', sep = ',')
velX_normalized = velX_normalized.sort_index(ascending=False)
velX_normalized.set_index('y_coord', inplace=True)

velY_normalized = pd.read_csv('velocity_Y_Normalized_0.txt', sep = ',')
velY_normalized = velY_normalized.sort_index(ascending=False)
velY_normalized.set_index('y_coord', inplace=True)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))
fig.tight_layout(h_pad = 5.0)

colour='turbo'

# viridis colour palette doesnt look too bad either - used to use coolwarm before

sns.heatmap(velX, cmap=colour, ax = ax[0, 0])
# ax[0, 0].set_title('Velocity (x component)')
ax[0, 0].annotate('a)', xy = (0.5, -0.2), xycoords='axes fraction')

sns.heatmap(velY, cmap=colour, ax = ax[0, 1])
# ax[0, 1].set_title('Velocity (y component)')
ax[0, 1].annotate('b)', xy = (0.5, -0.2), xycoords='axes fraction')

sns.heatmap(p, cmap=colour, ax = ax[1, 0])
# ax[1, 0].set_title('Pressure')
ax[1, 0].annotate('c)', xy = (0.5, -0.2), xycoords='axes fraction')

# sns.heatmap(p, cmap=colour, ax = ax[1, 1])
# ax[1, 1].quiver(velX, velY, scale=4)
# # ax[1, 1].set_title('Velocity vectors and pressure')
# ax[1, 1].annotate('d)', xy = (0.5, -0.2), xycoords='axes fraction')

sns.heatmap(p, cmap=colour, ax = ax[1, 1])
ax[1, 1].quiver(velX_normalized, velY_normalized, scale=30)
# ax[1, 1].set_title('Velocity vectors and pressure')
ax[1, 1].annotate('d)', xy = (0.5, -0.2), xycoords='axes fraction')

for ax in ax.flat:
    ax.set(xlabel='x (column number)')
    ax.set(ylabel='y (row number)')
    ax.set(aspect='equal')
    ax.tick_params(axis='x', rotation=0, labelsize=8)
    ax.tick_params(axis='y', rotation=0, labelsize=8)
```
